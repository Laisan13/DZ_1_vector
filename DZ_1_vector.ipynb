# Домашнее задание "Линейная алгебра. Векторы"

Даны вектора x = (1,1), y = (2, 0) и z = (0,2)

Надо:

Изобразить вектора на экране

Изобразить точку x + y + z

Найти угол между векторами x и y

```python
# Изобразить вектора на экране
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA

x = np.array([1 , 1])
y = np.array([2 , 0])
z = np.array([0 , 2])

plt.figure(figsize=(12, 7))
plt.plot(x[0], x[1], 'o')
plt.plot(y[0], y[1], 'o')
plt.plot(z[0], z[1], 'o')
plt.grid()
```

```python
# Изобразить точку x + y + z
c = x + y + z
print(c)
plt.figure(figsize=(12, 7))
plt.plot(c[0], c[1], 'o')
plt.grid()
```

```python
# Найти угол между векторами x и y
np.dot(x,y) #находим склярное произведение
LA.norm(x) #находим длину 
LA.norm(y)
cosangle = np.dot(x,y) / (LA.norm(x) * LA.norm(y)) #находим косинус угла
print ('Угол между векторами x и y:', np.around (np.rad2deg(np.arccos(cosangle))), 'градусов')
```
